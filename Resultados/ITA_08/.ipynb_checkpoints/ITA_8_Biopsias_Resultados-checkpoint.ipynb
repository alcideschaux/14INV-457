```python
df = pd.read_pickle('../Datos/ITA_8.pkl')
```

```python
df.shape
```

# Biopsia de próstata

## Realización de biopsia de próstata

```python
df['bx'].value_counts()
```

```python
df['bx'].value_counts(normalize=True).round(2)
```

```python
sns.countplot(x='bx', data=df)
plt.xlabel('Realización de biopsia de próstata')
plt.ylabel('Núm. pacientes')
plt.show()
```

## Categoría diagnóstica

```python
df['bx_categoria'] = df['bx_categoria'].astype('category')
df['bx_categoria'].cat.reorder_categories(
    ['Proceso inflamatorio/hiperplásico','Proceso preneoplásico','Proceso neoplásico maligno'],
    ordered=True,
    inplace=True
)
```

```python
df['bx_categoria'].value_counts(sort=False)
```

```python
df['bx_categoria'].value_counts(sort=False, normalize=True).round(2)
```

```python
sns.countplot(x='bx_categoria', data=df)
plt.xlabel('Categoría diagnóstica de biopsia de próstata')
plt.ylabel('Núm. pacientes')
plt.show()
```
